¿Qué es un SSID?
El SSID (Service Set IDentifier) o identificador de paquetes de servicio es el nombre que identifica una red inalámbrica WIFI y el que viaja junto con cada paquete de información de la misma, de forma que pueda ser identificado como parte de ella.
